```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("Sample.jpg")
img.shape
```

```python
plt.matshow(img)
```

```python
person = img[650:800,150:200]
```

```python
img[0:150,0:50] = person
```

```python
b,g,r = cv2.split(img)
```

```python
b = np.zeros(b.shape)
np.where(b>200,0,255)
img = cv2.merge((r,g,b))
```

```python
cv2.imshow("Image",img)
cv2.waitKey(0) 
cv2.destroyAllWindows()
```
